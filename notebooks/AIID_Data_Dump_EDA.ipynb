```python
import bson
```

```python
def load_bson(loc):
    with open(loc, 'rb') as f:
        data = bson.decode_all(f.read())
        return pd.DataFrame(data)
```

```python
incidents_df = load_bson('aiidprod/incidents.bson')
```

```python
candidates_df = load_bson('aiidprod/candidates.bson')
```
